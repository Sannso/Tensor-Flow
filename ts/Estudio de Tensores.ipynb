## Aprendizaje supervisado
 Un algoritmo de aprendizaje automatico, contruye un modelo, al examinar muchos ejemplos e intenta encontrar un modelo que minimize la perdida
###### Este proceso se llama Minimizacion Empirica de Riesgos

La perdida es la penalizacion por una mala prediccion, la perdida es un numero que indica cuan mala fue la prediccion del modelo en un solo ejempolo, si la prediccion es perfecta la perdida es 0



Los modelos de regresion lineal como los que estamos viendo usan una funcion de perdida llamada perdida cuadrada o perdida de L_2

Que es el cuadrado de la diferencia entre la etiqueta y la prediccion: 



Pero con esta formula solo calculamos la perdida para cada ejemplo y lo que nos interesa es calcular la perdida promedio de todas las predicciones.

A partir del calculo de perdida cuadrada el modelo calcula la perdida cuadractica promedio mediante la siguiente formula: 

###### FORMULA: ERROR CUADRATICO MEDIO
se trata de un modelo estadistico que nos permite calcular el promedio de error que es lo que nos interesa




# Reduccion de perdida

La clave del aprendizaje automatico consiste en reducir lo maximo posible la perdida y se hace mediante un metodo iterativo, esta iteracion se llevara a cabo hasta que se reduzca esa perdida lo maximo posible.




En un modelo simplificado que tome una sola caracteristica y devuelva una prediccion la funcion de prediccion tendria una formula como esta:
$y' = b + w_1x_1$

La iteracion se repite hasta que la perdida general deja de cambiar o cambia muy lentamente, cuando eso sucesede se dice que el modelo ha convergido

En un problema de regresion como este la grafica resultante siempre sera convexa con forma de cuenco, porque los problemas de regresion solo tienen un minimo, solo tienen un lugar donde la pendiente es exactamente 0 y ese minimo es donde converge la funcion de perdida.

calcular esa funcion de perdida para cada valor posible de $W_1$ y para todos los conjuntos de datos que utilicemos es muy poco eficiente

Por eso <b>TensorFlow</b>, <b>utiliza un mecanismo llamado Reduccion de perdida - Descenso de Gradiente</b>

El valor o el salto que el punto da en el descenso del gradiente depende del programador y a ese salto se le llama <b>tasa de aprendizaje</b>

```python
# Ejemplo de la funcion random_normal()

import tensorflow as tf

tensor_random = tf.random_normal([3, 3], dtype=tf.float64)

with tf.Session() as sess:
    print(sess.run(tensor_random))
```

```python
# Ejemplo de la funcion reshape()

import tensorflow as tf

tensorDim1 = tf.constant([1., 2., 3., 4.])
mensaje = tf.string_join(["Tensor redimensionado: "])
tensorDim2 = tf.reshape(tensorDim1, [2, 2])

with tf.Session() as sess:
    print(sess.run(tensorDim1))
    print(sess.run(mensaje))
    print(sess.run(tensorDim2))
```

```python
# Ejemplo de la funcion slice()
 
import tensorflow as tf

tensorCompleto = tf.constant([ [1, 2, 3], [4, 5, 6], [7, 8, 9] ])
mensaje = tf.string_join(["Tensor recortado: "])
tensor_slice = tf.slice(tensorCompleto, [2, 0], [1, 3])
# tf.slice(tensor, coordenadas donde comienza, dimension a la que se ajusta)

with tf.Session() as sess:
    print(sess.run(tensorCompleto))
    print(sess.run(mensaje))
    print(sess.run(tensor_slice))
```

```python
# Ejemplo de la funcion stack()

import tensorflow as tf

tensor1 = tf.constant( [1, 2] )
tensor2 = tf.constant( [9, 8] )
tensor3 = tf.constant( [7, 6] )
tensor_stack = tf.stack([tensor1, tensor2, tensor3])
# Stack devuelve un tensor de la dimension de los tensores + 1 (n+1)
# Los tensores deben que tener el mismo rango y la misma forma

mensaje = tf.string_join([" ----- Mensaje de separacion ----- "])
mensaje2 = tf.string_join([" ----- Dimencion de stack: "])
shape_stack = tf.shape(tensor_stack)
rank_stack = tf.rank(tensor_stack)

with tf.Session() as sess:
    print(sess.run(tensor1))
    print(sess.run(mensaje))
    print(sess.run(tensor2))
    print(sess.run(mensaje))
    print(sess.run(tensor3))
    print(sess.run(mensaje))
    print(sess.run(tensor_stack))
    print(sess.run(mensaje2))
    print(sess.run(shape_stack))
    print(sess.run(mensaje))
    print(sess.run(rank_stack))
```

```python
# Ejemplo de la funcion unstack

import tensorflow as tf

tensor1 = tf.constant( [1, 2] )
tensor2 = tf.constant( [9, 8] )
tensor3 = tf.constant( [7, 6] )
tensor_stack = tf.stack([tensor1, tensor2, tensor3])

tensor_unstack = tf.unstack(tensor_stack)

with tf.Session() as sess:
    print(sess.run(tensor_unstack))
```

```python
# Ejemplos de operaciones matematicas

import tensorflow as tf

tensor1 = tf.constant([1., 3.])
tensor2 = tf.constant([4., 2.])

suma = tf.add(tensor1, tensor2)

resta = tf.subtract(tensor1, tensor2)

producto = tf.multiply(tensor1, tensor2)

division = tf.divide(tensor1, tensor2)

with tf.Session() as sess:
    print(sess.run(suma))
    print(sess.run(resta))
    print(sess.run(producto))
    print(sess.run(division))

```
